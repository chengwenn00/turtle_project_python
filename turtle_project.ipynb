```python
import turtle
from turtle import Screen, Turtle, Shape
import math
from math import sqrt

#This section is the global variables section, to be changed by the user inputs later
CURSOR_SIZE = 20
FONT_SIZE = 12
FONT = ('Arial', FONT_SIZE, 'bold')
pencolor = "black"
fillcolor = '#FFA500'
x_coord_list = []
y_coord_list = []
screen = Screen()
shape = Shape('compound')



#This function adjusts the polygon display settings like pen color and fill color
def polygon_display(x, y):
    global pencolor
    global fillcolor
    
    select_pen_color = int(turtle.numinput("Select pen color","1 - black, 2 - blue, 3 - red",
                          default=1,
                          minval=1,
                          maxval=3
                          ))
    if select_pen_color == 1:
        pencolor = "black"
    elif select_pen_color == 2:
        pencolor = "blue"
    else:
        pencolor = "red"
    
    select_fill_color = turtle.textinput("Select fill color","#xxxxxx")
    fillcolor = select_fill_color
    turtle.TurtleScreen._RUNNING = True

#This function draws the polygon
def draw_polygon(x, y):
    turtle.clear()
    turtle.up()
    turtle.setpos(0,0)
    turtle.down()
    global x_coord_list
    global y_coord_list
    x_coord_list = [0]
    y_coord_list = [0]
    turtle.pencolor(pencolor)
    turtle.fillcolor(fillcolor)
    turtle.begin_fill()
    
    
    
    #This section asks the user how many vertices there should be
    num_vertices = int(turtle.numinput("Vertices count","Key in number of vertices (between 3 and 10)",
                          default=3,
                          minval=3,
                          maxval=10
                          ))

    #This section initialises the x and y coordinates for each number of vertices and connects them
    shape = Shape('compound')
    turtle.begin_poly()
    for i in range(num_vertices-1):
        
        
        x_coord = int(turtle.numinput("Coordinates","Key in X-coordinate (between -300 and 300!)",
                          default=0,
                          minval=-300,
                          maxval=300
                          ))
        y_coord = int(turtle.numinput("Coordinates","Key in Y-coordinate (between -300 and 300!)",
                          default=0,
                          minval=-300,
                          maxval=300
                          ))
        x_coord_list.append(x_coord)
        y_coord_list.append(y_coord)
        turtle.goto(x_coord, y_coord)
        
        #This part checks if this line intersects with any other line
        if i > 1:
            for j in range(i-1):
                p1 = [x_coord_list[j],y_coord_list[j]]
                q1 = [x_coord_list[j+1],y_coord_list[j+1]]
                p2 = [x_coord_list[i],y_coord_list[i]]
                q2 = [x_coord_list[i+1],y_coord_list[i+1]]
                
                if doIntersect(p1, q1, p2, q2):
                    turtle.clear()
                    turtle.write("Self-Intersection detected! Draw a new polygon again.")
                else:
                    continue
    turtle.goto(0,0)
    turtle.end_fill()
    turtle.end_poly()
    shape.addcomponent(turtle.get_poly(), fillcolor, pencolor)
    screen.register_shape('polygon', shape)
    turtle.TurtleScreen._RUNNING = True

#This function checks if point q lies on line segment 'pr' given three collinear points p, q, r
def onSegment(p, q, r):
    if ( (q[0] <= max(p[0], r[0])) and (q[0] >= min(p[0], r[0])) and
           (q[1] <= max(p[1], r[1])) and (q[1] >= min(p[1], r[1]))):
        return True
    return False

#This function finds the orientation of the segments
def orientation(p, q, r):
    # to find the orientation of an ordered triplet (p,q,r)
    # function returns the following values:
    # 0 : Collinear points
    # 1 : Clockwise points
    # 2 : Counterclockwise
    
   
    val = (float(q[1] - p[1]) * (r[0] - q[0])) - (float(q[0] - p[0]) * (r[1] - q[1]))
    if (val > 0):
         
        # Clockwise orientation
        return 1
    elif (val < 0):
         
        # Counterclockwise orientation
        return 2
    else:
         
        # Collinear orientation
        return 0

#This function returns true if the line segment 'p1q1' and 'p2q2' intersect.
def doIntersect(p1,q1,p2,q2):
     
    # Find the 4 orientations required for
    # the general and special cases
    o1 = orientation(p1, q1, p2)
    o2 = orientation(p1, q1, q2)
    o3 = orientation(p2, q2, p1)
    o4 = orientation(p2, q2, q1)
 
    # General case
    if ((o1 != o2) and (o3 != o4)):
        return True
 
    # Special Cases
 
    # p1 , q1 and p2 are collinear and p2 lies on segment p1q1
    if ((o1 == 0) and onSegment(p1, p2, q1)):
        return True
 
    # p1 , q1 and q2 are collinear and q2 lies on segment p1q1
    if ((o2 == 0) and onSegment(p1, q2, q1)):
        return True
 
    # p2 , q2 and p1 are collinear and p1 lies on segment p2q2
    if ((o3 == 0) and onSegment(p2, p1, q2)):
        return True
 
    # p2 , q2 and q1 are collinear and q1 lies on segment p2q2
    if ((o4 == 0) and onSegment(p2, q1, q2)):
        return True
 
    # If none of the cases
    return False


#This function allows users to add or remove a vertice and redraws the new polygon.
def vert_change(x,y):
    print("Here are the x-coordinates for vertices")
    print(x_coord_list)
    print("Here are the y-coordinates for vertices")
    print(y_coord_list)
    add_or_remove = int(turtle.numinput("Add or Remove","Key in 1 to add new vertice, 2 to remove existing vertice",
                          default=1,
                          minval=1,
                          maxval=2
                          ))
    if add_or_remove == 1:
        add_x = int(turtle.numinput("Coordinates","Key in X-coordinate (between -300 and 300!)",
                          default=0,
                          minval=-300,
                          maxval=300
                          ))
        add_y = int(turtle.numinput("Coordinates","Key in Y-coordinate (between -300 and 300!)",
                          default=0,
                          minval=-300,
                          maxval=300
                          ))
        
        x_coord_list.append(add_x)
        y_coord_list.append(add_y)
        turtle.clear()
        
        turtle.up()
        turtle.setpos(0,0)
        turtle.down()
        shape = Shape('compound')
        
        turtle.begin_fill()
        turtle.begin_poly()
        
        for i in range(len(x_coord_list)-1):
            turtle.goto(x_coord_list[i+1], y_coord_list[i+1])
        
        #This part checks if this line intersects with any other line
            if i > 1:
                for j in range(i-1):
                    p1 = [x_coord_list[j],y_coord_list[j]]
                    q1 = [x_coord_list[j+1],y_coord_list[j+1]]
                    p2 = [x_coord_list[i],y_coord_list[i]]
                    q2 = [x_coord_list[i+1],y_coord_list[i+1]]
                
                    if doIntersect(p1, q1, p2, q2):
                        turtle.clear()
                        turtle.write("Self-Intersection detected! Draw a new polygon again.")
                    else:
                        continue
        turtle.goto(0,0)
        turtle.end_fill()
        turtle.end_poly()
        shape.addcomponent(turtle.get_poly(), fillcolor, pencolor)
        screen.register_shape('polygon', shape)
        turtle.TurtleScreen._RUNNING = True
        
                        
    if add_or_remove == 2:
        if len(x_coord_list) <= 3:
            turtle.clear()
            turtle.write("There are too little vertices to remove!")
            
        else:
            remove_vertice = int(turtle.numinput("Coordinates","Key in vertice number excluding origin (e.g. 1, 2, 3)",
                          default=1,
                          minval=1,
                          maxval=len(x_coord_list)-1
                          ))
            
            del x_coord_list[remove_vertice]
            del y_coord_list[remove_vertice]
            turtle.clear()
            turtle.begin_fill()
            shape = Shape('compound')
            turtle.begin_poly()
            
            for i in range(len(x_coord_list)-1):
                turtle.goto(x_coord_list[i+1], y_coord_list[i+1])

                #This part checks if this line intersects with any other line
                if i > 1:
                    for j in range(i-1):
                        p1 = [x_coord_list[j],y_coord_list[j]]
                        q1 = [x_coord_list[j+1],y_coord_list[j+1]]
                        p2 = [x_coord_list[i],y_coord_list[i]]
                        q2 = [x_coord_list[i+1],y_coord_list[i+1]]

                        if doIntersect(p1, q1, p2, q2):
                            turtle.clear()
                            turtle.write("Self-Intersection detected! Draw a new polygon again.")
                        else:
                            continue
            turtle.goto(0,0)
            turtle.end_fill()
            turtle.end_poly()
            shape.addcomponent(turtle.get_poly(), fillcolor, pencolor)
            screen.register_shape('polygon', shape)
            turtle.TurtleScreen._RUNNING = True

            
            
#This function rotates the polygon
def rotate_polygon(x,y):
    turtle.clear()
    turtle.shape('polygon')

    rotate_deg = int(turtle.numinput("Rotation Degree","Key in degree to rotate by (between 0 and 360 degrees)",
                          default=180,
                          minval=0,
                          maxval=360
                          ))
    turtle.left(rotate_deg)

#This function moves the polygon
def move_polygon(x,y):
    turtle.clear()
    turtle.shape('polygon')

    direction_angle = int(turtle.numinput("Direction","Key in direction to face (between 0 and 360 degrees)",
                          default=180,
                          minval=0,
                          maxval=360
                          ))
    turtle.setheading(direction_angle)
    
    x_coord = int(turtle.numinput("Coordinates","Key in X-coordinate (between -300 and 300!)",
                          default=180,
                          minval=-300,
                          maxval=300
                          ))
    
    y_coord = int(turtle.numinput("Coordinates","Key in Y-coordinate (between -300 and 300!)",
                          default=180,
                          minval=-300,
                          maxval=300
                          ))
    turtle.up()
    turtle.goto(x_coord,y_coord)

#This function calculates the perimeter of the polygon
def perimeter_polygon(x,y):
    len_polygon = 0
    if len(x_coord_list) < 3:
        turtle.write("There is no polygon!")
    else:
        for i in range(len(x_coord_list)):
            if i < (len(x_coord_list) - 1):
                len_polygon += sqrt((x_coord_list[i+1] - x_coord_list[i])**2 + (y_coord_list[i+1] - y_coord_list[i])**2)
            elif i == (len(x_coord_list) - 1):
                len_polygon += sqrt((x_coord_list[i] - x_coord_list[0])**2 + (y_coord_list[i] - y_coord_list[0])**2)
    turtle.hideturtle()
    turtle.up()
    turtle.setpos(-400, 220)
    turtle.write("Perimeter: ")
    turtle.up()
    turtle.setpos(-400, 200)
    turtle.write(len_polygon)

#This function calculates the area of the polygon
def area_polygon(x,y):
    area_polygon = 0
    if len(x_coord_list) < 3:
        turtle.write("There is no polygon!")
    else:
        for i in range(len(x_coord_list)):
            if i < (len(x_coord_list) - 1):
                area_polygon += x_coord_list[i]*y_coord_list[i+1] - y_coord_list[i]*x_coord_list[i+1]
            elif i == (len(x_coord_list) - 1):
                area_polygon += x_coord_list[i]*y_coord_list[0] - y_coord_list[i]*x_coord_list[0]
    area_polygon = area_polygon/2
    
    turtle.hideturtle()
    turtle.up()
    turtle.setpos(-400, 180)
    turtle.write("Area: ")
    turtle.up()
    turtle.setpos(-400, 160)
    turtle.write(area_polygon)

    
# def resize_polygon(x,y):
#     resize_factor = int(turtle.numinput("Resize","Key in resizing factor (1.0 being same size)",
#                           default=1,
#                           minval=0.1,
#                           maxval=5
#                           ))
    
#     turtle.shapesize(stretch_wid=resize_factor, stretch_len=resize_factor, outline=None)
            
            
#This section draws the Draw a Polygon! button. This draws a 3-sided polygon by default (minimum sides)
#This section calls the function draw_polygon
draw = Turtle()
draw.speed(speed=10)
draw.hideturtle()
draw.shape('circle')
draw.fillcolor('red')
draw.penup()
draw.goto(-350, 350)
draw.write("Draw a polygon!", align='center', font=FONT)
draw.sety(350 + CURSOR_SIZE + FONT_SIZE)

draw.onclick(draw_polygon)
draw.showturtle()

#This section draws the Polygon Display button. This allows users to select colours for the polygon
#This section calls the function polygon_display
poly_disp = Turtle()
poly_disp.speed(speed=10)
poly_disp.hideturtle()
poly_disp.shape('circle')
poly_disp.fillcolor('red')
poly_disp.penup()
poly_disp.goto(-200, 350)
poly_disp.write("Polygon Display", align='center', font=FONT)
poly_disp.sety(350 + CURSOR_SIZE + FONT_SIZE)

poly_disp.onclick(polygon_display)
poly_disp.showturtle()

#This section displays the coordinates of the vertices inputted into the program
vert_list = Turtle()
vert_list.speed(speed=10)
vert_list.hideturtle()
vert_list.shape('circle')
vert_list.fillcolor('red')
vert_list.penup()
vert_list.goto(-40, 350)
vert_list.write("Add/Remove Vertices", align='center', font=FONT)
vert_list.sety(350 + CURSOR_SIZE + FONT_SIZE)

vert_list.onclick(vert_change)
vert_list.showturtle()

#This section allows for rotation of polygon
rotate = Turtle()
rotate.speed(speed=10)
rotate.hideturtle()
rotate.shape('circle')
rotate.fillcolor('red')
rotate.penup()
rotate.goto(110, 350)
rotate.write("Rotate Polygon", align='center', font=FONT)
rotate.sety(350 + CURSOR_SIZE + FONT_SIZE)

rotate.onclick(rotate_polygon)
rotate.showturtle()


#This section allows for movement of polygon
movement = Turtle()
movement.speed(speed=10)
movement.hideturtle()
movement.shape('circle')
movement.fillcolor('red')
movement.penup()
movement.goto(260, 350)
movement.write("Move Polygon", align='center', font=FONT)
movement.sety(350 + CURSOR_SIZE + FONT_SIZE)

movement.onclick(move_polygon)
movement.showturtle()


#This section calculates the perimeter of the polygon
perimeter = Turtle()
perimeter.speed(speed=10)
perimeter.hideturtle()
perimeter.shape('circle')
perimeter.fillcolor('red')
perimeter.penup()
perimeter.goto(-350, 300)
perimeter.write("Perimeter of Polygon", align='center', font=FONT)
perimeter.sety(300 + CURSOR_SIZE + FONT_SIZE)

perimeter.onclick(perimeter_polygon)
perimeter.showturtle()

#This section calculates the area of the polygon
area = Turtle()
area.speed(speed=10)
area.hideturtle()
area.shape('circle')
area.fillcolor('red')
area.penup()
area.goto(-200, 300)
area.write("Area of Polygon", align='center', font=FONT)
area.sety(300 + CURSOR_SIZE + FONT_SIZE)

area.onclick(move_polygon)
area.showturtle()


#This section resizes the polygon
# resize_poly = Turtle()
# resize_poly.speed(speed=10)
# resize_poly.hideturtle()
# resize_poly.shape('circle')
# resize_poly.fillcolor('red')
# resize_poly.penup()
# resize_poly.goto(-50, 300)
# resize_poly.write("Resizing Polygon", align='center', font=FONT)
# resize_poly.sety(300 + CURSOR_SIZE + FONT_SIZE)

# resize_poly.onclick(resize_polygon)
# resize_poly.showturtle()




screen = Screen()
screen.mainloop()

turtle.done()
```
